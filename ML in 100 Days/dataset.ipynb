```python
import seaborn as sns

categories = df['Category'].unique()
category_colors = plt.cm.tab20c.colors[:len(categories)]

# Group and aggregate data to handle duplicates
agg_data = df.groupby(['Month Year', 'Category'])['order_quantity'].sum().reset_index()

# Pivot the aggregated data
pivot_table = agg_data.pivot(index='Month Year', columns='Category', values='order_quantity')

# Create a stacked bar plot
pivot_table.plot(kind='bar', stacked=True, figsize=(12, 6), color=[category_colors[i % len(category_colors)] for i in range(len(categories))])

plt.xlabel('Month/Year')
plt.ylabel('Order Quantity')
plt.title('Category Sales Over Time')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')

# Line Chart for Time Series Analysis
plt.figure(figsize=(12, 6))
for category in categories:
    subset = agg_data[agg_data['Category'] == category]
    plt.plot(subset['Month Year'], subset['order_quantity'], label=category)

plt.title('Time Series Analysis - Category Sales')
plt.xlabel('Month/Year')
plt.ylabel('Order Quantity')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Pie Chart for Market Share
plt.figure(figsize=(20, 20))
pie_chart = df.groupby('Category')['order_quantity'].sum().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Market Share by Category')
plt.ylabel('')
plt.show()

# Box Plot for SKU Performance Comparison
plt.figure(figsize=(15, 8))
box_plot = sns.boxplot(x='Category', y='order_quantity', data=df, palette='pastel')
box_plot.set_xticklabels(box_plot.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.title('SKU Performance Comparison')
plt.xlabel('Category')
plt.ylabel('Order Quantity')

# Set y-axis ticks with a difference of 200
plt.yticks(range(0, int(df['order_quantity'].max()) + 100, 100))

plt.show()
```

```python
payment_mode_counts = df['payment_mode'].value_counts()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(payment_mode_counts, labels=payment_mode_counts.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)
plt.title('Payment Mode Distribution')
plt.show()
```

```python
# Calculate the share of each brand
brand_counts = df['Brand'].value_counts(normalize=True) * 100

# Filter brands with more than 1% market share
selected_brands = brand_counts[brand_counts > 1]

# Get categories for selected brands
selected_categories = df[df['Brand'].isin(selected_brands.index)]['Category'].unique()

# Plot a pie chart for the selected brands with categories
plt.figure(figsize=(8, 8))
selected_brands.plot.pie(autopct='%1.1f%%', startangle=90, colors=plt.cm.Set3.colors)
plt.title('Share of Brands with >1% Market Share')
plt.ylabel('')
plt.legend(selected_categories, bbox_to_anchor=(1, 0.5), loc="center left", title='Categories')
plt.show()
```
