# **Working with Dates**

```python
# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

**1. Extract Year**

```python
date['date_year'] = date['date'].dt.year
date.sample(5)
```

**2. Extract Month**

```python
date['date_month'] = date['date'].dt.month
date.head()
```

**3. Extract name of the month**

```python
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

**4. Extract Days**

```python
date['date_day'] = date['date'].dt.day
date.head()
```

**5. Day of the week**

```python
date['date_dow'] = date['date'].dt.dayofweek
date.head()
```

**6. Name of the Day**

```python
date['date_day_name'] = date['date'].dt.day_name()
date.head()
```

**7. Is weekend?**

```python
date['date_weekend'] = np.where(date['date_day_name'].isin(['Sunday', 'Saturday']), 1,0)
date.drop(columns=['product_id','city_id','orders'])
```

**8. Extract week of the year**

```python
date['date_week'] = date['date'].dt.isocalendar().week 
date.drop(columns=['product_id','city_id','orders']).head()
```

**9. Extract Quarter**

```python
date['quarter'] = date['date'].dt.quarter
date.drop(columns=['product_id','city_id','orders']).head()
```

**10. Extract Semester**

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

**11. Extract Time elapsed between dates**

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed

np.round((today -date['date']) / np.timedelta64(1, 'M'),0)
```
